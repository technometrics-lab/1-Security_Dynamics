```python
import matplotlib.pyplot as plt

%matplotlib inline
import numpy as np
import math
```

```python
def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a
```

```python
x = np.arange(-6, 6, 0.2)
```

```python
sig = sigmoid(x)
```

```python
plt.plot(x, sig, color='red')
```

```python
fig, ax = plt.subplots()

plt.plot(x, sig, color='red')
plt.vlines(x=0, ymin=0, ymax=0.5, colors='green', ls=':')
plt.ylabel("Technology diffusion")
plt.yticks([0])
plt.xlabel("Time")
plt.xticks([])
plt.margins(x=0, y=0)
#plt.title("S-curve of technological development")

textstr_1 = 'maturity'
textstr_2 = 'growth'
textstr_3 = 'introduction'
textstr_4 = 'inflection point'

# these are matplotlib.patch.Patch properties
props = dict(facecolor='white', alpha=0.5)

# place a text box in upper left in axes coords
ax.text(0.55, 0.95, textstr_1, transform=ax.transAxes, fontsize=10,
        verticalalignment='top')
ax.text(0.37, 0.50, textstr_2, transform=ax.transAxes, fontsize=10,
        verticalalignment='top')
ax.text(0.05, 0.15, textstr_3, transform=ax.transAxes, fontsize=10,
        verticalalignment='top')
ax.text(0.52, 0.52, textstr_4, transform=ax.transAxes, fontsize=8,
        verticalalignment='top')
plt.savefig("scurve.pdf")
plt.show()
```
